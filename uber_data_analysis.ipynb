After identifying zero-distance trips, I quantified their business impact. I found that although they were only ~2.6% of trips, they still generated ~2.4% of total revenue. I then translated this into operational recommendations like GPS audits and minimum-fare review.

```python
# Business impact summary
print(f"Zero distance trips %: {len(zero_distance)/len(df)*100:.2f}%")
print(f"Zero distance revenue %: {zero_distance['total_amount'].sum() / df['total_amount'].sum() * 100:.2f}%")
print("\nRecommendations:")
print("- Investigate GPS/meter issues in top locations (132, 265, etc.)")
print("- Consider minimum fare review for very short trips")
```

1 Distribution of Trip Distance

This histogram shows how trip distances are distributed, revealing that a large number of trips are recorded as zero miles, which signals potential GPS or data quality issues.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.hist(df['trip_distance'], bins=50)
plt.title("Distribution of Trip Distance")
plt.xlabel("Miles")
plt.ylabel("Number of Trips")
plt.show()
```

This scatter plot shows a strong positive relationship between distance and fare, but it also reveals extreme outliers where trips with zero recorded distance were billed very high amounts.

```python
plt.figure(figsize=(8,5))
plt.scatter(df['trip_distance'], df['fare_amount'], alpha=0.3)
plt.title("Trip Distance vs Fare")
plt.xlabel("Trip Distance (miles)")
plt.ylabel("Fare Amount ($)")
plt.show()
```

This histogram shows that zero-distance trips include a wide range of fares, including extremely high values, confirming that these were real, fully billed rides rather than just cancellations.

```python
plt.figure(figsize=(8,5))
plt.hist(zero_distance['fare_amount'], bins=50)
plt.title("Fare Distribution for Zero-Distance Trips")
plt.xlabel("Fare ($)")
plt.ylabel("Trips")
plt.show()
```

This boxplot compares fare distributions and shows that zero-distance trips have unusually high and more volatile charges compared to normal trips.

```python
plt.figure(figsize=(8,5))
plt.boxplot(
    [df['fare_amount'], zero_distance['fare_amount']],
    labels=['All Trips','Zero-Distance Trips']
)
plt.title("Fare Comparison")
plt.ylabel("Fare ($)")
plt.show()
```

This chart shows that although zero-distance rides are a small fraction of trips, they still contribute over 2% of total revenue, making them financially significant

```python
labels = ['Normal Trips','Zero-Distance Trips']
values = [
    df['total_amount'].sum() - zero_distance['total_amount'].sum(),
    zero_distance['total_amount'].sum()
]

plt.figure(figsize=(6,6))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Revenue Share from Zero-Distance Trips")
plt.show()
```
